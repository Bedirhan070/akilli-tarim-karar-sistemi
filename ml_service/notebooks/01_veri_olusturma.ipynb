```python
# Her sehir icin hava verisi cek
tum_hava_verisi = []

for _, satir in sehir_df.iterrows():
    print(f"{satir['sehir']} icin veri cekiliyor...")
    
    url = "https://archive-api.open-meteo.com/v1/archive"
    params = {
        "latitude": satir["lat"],
        "longitude": satir["lon"],
        "start_date": "2022-01-01",
        "end_date": "2024-12-31",
        "daily": [
            "temperature_2m_max",
            "temperature_2m_min",
            "precipitation_sum",
            "windspeed_10m_max"
        ],
        "timezone": "Europe/Istanbul"
    }
    
    response = requests.get(url, params=params)
    data = response.json()
    
    df_temp = pd.DataFrame(data["daily"])
    df_temp["sehir"] = satir["sehir"]
    df_temp["bolge"] = satir["bolge"]
    
    tum_hava_verisi.append(df_temp)

hava_df = pd.concat(tum_hava_verisi, ignore_index=True)
hava_df.rename(columns={
    "time": "tarih",
    "temperature_2m_max": "max_sicaklik",
    "temperature_2m_min": "min_sicaklik",
    "precipitation_sum": "yagis",
    "windspeed_10m_max": "ruzgar_hizi"
}, inplace=True)

print(f"\nToplam satir: {len(hava_df)}")
print(f"Sutunlar: {hava_df.columns.tolist()}")
print(hava_df.head())
```

```python
# Veriyi kaydet
hava_df.to_csv("../data/turkiye_hava_verisi.csv", index=False)
sehir_df.to_csv("../data/sehir_listesi.csv", index=False)

print("Kaydedildi!")
print(f"  → data/turkiye_hava_verisi.csv  ({len(hava_df)} satir)")
print(f"  → data/sehir_listesi.csv        ({len(sehir_df)} satir)")
```

```python
import pandas as pd
import requests

sehir_df = pd.read_csv("/Users/kasimozel/Desktop/Akıllı Tarım Asistanı/akilli-tarim-karar-sistemi/ml_service/data/sehir_listesi.csv")

tum_hava_verisi = []

for _, satir in sehir_df.iterrows():
    print(f"{satir['sehir']} icin veri cekiliyor...")
    
    url = "https://archive-api.open-meteo.com/v1/archive"
    params = {
        "latitude": satir["lat"],
        "longitude": satir["lon"],
        "start_date": "2022-01-01",
        "end_date": "2024-12-31",
        "daily": [
            "temperature_2m_max",
            "temperature_2m_min",
            "precipitation_sum",
            "windspeed_10m_max",
            "relative_humidity_2m_max"
        ],
        "timezone": "Europe/Istanbul"
    }
    
    response = requests.get(url, params=params)
    data = response.json()
    
    if "daily" not in data:
        print(f"HATA - {satir['sehir']}: {data}")
        continue
    
    df_temp = pd.DataFrame(data["daily"])
    df_temp["sehir"] = satir["sehir"]
    df_temp["bolge"] = satir["bolge"]
    tum_hava_verisi.append(df_temp)

hava_df = pd.concat(tum_hava_verisi, ignore_index=True)
hava_df.rename(columns={
    "time"                    : "tarih",
    "temperature_2m_max"      : "max_sicaklik",
    "temperature_2m_min"      : "min_sicaklik",
    "precipitation_sum"       : "yagis",
    "windspeed_10m_max"       : "ruzgar_hizi",
    "relative_humidity_2m_max": "nem"
}, inplace=True)

print(f"\nToplam satir: {len(hava_df)}")
print(hava_df.head())

DATA_PATH = "/Users/kasimozel/Desktop/Akıllı Tarım Asistanı/akilli-tarim-karar-sistemi/ml_service/data/"
hava_df.to_csv(DATA_PATH + "turkiye_hava_verisi.csv", index=False)
print("\nKaydedildi!")
```

```python
import time
time.sleep(60)  # 1 dakika bekle

eksik_sehirler = ["Edirne", "Adana", "Samsun", "Tokat"]
eksik_df = sehir_df[sehir_df["sehir"].isin(eksik_sehirler)]

eksik_veri = []

for _, satir in eksik_df.iterrows():
    print(f"{satir['sehir']} icin veri cekiliyor...")
    
    url = "https://archive-api.open-meteo.com/v1/archive"
    params = {
        "latitude": satir["lat"],
        "longitude": satir["lon"],
        "start_date": "2022-01-01",
        "end_date": "2024-12-31",
        "daily": [
            "temperature_2m_max",
            "temperature_2m_min",
            "precipitation_sum",
            "windspeed_10m_max",
            "relative_humidity_2m_max"
        ],
        "timezone": "Europe/Istanbul"
    }
    
    response = requests.get(url, params=params)
    data = response.json()
    
    if "daily" not in data:
        print(f"HATA - {satir['sehir']}: {data}")
        continue
    
    df_temp = pd.DataFrame(data["daily"])
    df_temp["sehir"] = satir["sehir"]
    df_temp["bolge"] = satir["bolge"]
    eksik_veri.append(df_temp)
    time.sleep(5)  # Her istek arasinda 5 saniye bekle

# Mevcut veriyle birlestir
eksik_hava_df = pd.concat(eksik_veri, ignore_index=True)
eksik_hava_df.rename(columns={
    "time"                    : "tarih",
    "temperature_2m_max"      : "max_sicaklik",
    "temperature_2m_min"      : "min_sicaklik",
    "precipitation_sum"       : "yagis",
    "windspeed_10m_max"       : "ruzgar_hizi",
    "relative_humidity_2m_max": "nem"
}, inplace=True)

hava_df = pd.concat([hava_df, eksik_hava_df], ignore_index=True)
hava_df = hava_df.sort_values(["sehir", "tarih"]).reset_index(drop=True)

print(f"Toplam satir: {len(hava_df)}")
print(f"Sehirler: {hava_df['sehir'].unique()}")

hava_df.to_csv(DATA_PATH + "turkiye_hava_verisi.csv", index=False)
print("Kaydedildi!")
```
