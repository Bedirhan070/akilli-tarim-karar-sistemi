```python
import pandas as pd
import numpy as np
import requests

# Turkiye'de bugdayin yetistigi sehirler ve bolgeler
sehirler = [
    # Bolge, Sehir, Enlem, Boylam
    {"bolge": "Ic Anadolu",          "sehir": "Konya",      "lat": 37.87, "lon": 32.49},
    {"bolge": "Ic Anadolu",          "sehir": "Ankara",     "lat": 39.92, "lon": 32.85},
    {"bolge": "Ic Anadolu",          "sehir": "Eskisehir",  "lat": 39.77, "lon": 30.52},
    {"bolge": "Ic Anadolu",          "sehir": "Kayseri",    "lat": 38.72, "lon": 35.49},
    {"bolge": "Ic Anadolu",          "sehir": "Sivas",      "lat": 39.74, "lon": 37.01},
    {"bolge": "Dogu Anadolu",        "sehir": "Erzurum",    "lat": 39.90, "lon": 41.27},
    {"bolge": "Dogu Anadolu",        "sehir": "Malatya",    "lat": 38.35, "lon": 38.31},
    {"bolge": "Dogu Anadolu",        "sehir": "Bitlis",     "lat": 38.40, "lon": 42.12},
    {"bolge": "Dogu Anadolu",        "sehir": "Van",        "lat": 38.49, "lon": 43.38},
    {"bolge": "Dogu Anadolu",        "sehir": "Agri",       "lat": 39.72, "lon": 43.05},
    {"bolge": "Guneydogu Anadolu",   "sehir": "Diyarbakir", "lat": 37.91, "lon": 40.23},
    {"bolge": "Guneydogu Anadolu",   "sehir": "Sanliurfa",  "lat": 37.16, "lon": 38.79},
    {"bolge": "Guneydogu Anadolu",   "sehir": "Gaziantep",  "lat": 37.06, "lon": 37.38},
    {"bolge": "Ege",                 "sehir": "Izmir",      "lat": 38.41, "lon": 27.13},
    {"bolge": "Ege",                 "sehir": "Afyon",      "lat": 38.75, "lon": 30.54},
    {"bolge": "Marmara",             "sehir": "Tekirdag",   "lat": 40.97, "lon": 27.51},
    {"bolge": "Marmara",             "sehir": "Edirne",     "lat": 41.67, "lon": 26.55},
    {"bolge": "Akdeniz",             "sehir": "Adana",      "lat": 37.00, "lon": 35.32},
    {"bolge": "Karadeniz",           "sehir": "Samsun",     "lat": 41.28, "lon": 36.33},
    {"bolge": "Karadeniz",           "sehir": "Tokat",      "lat": 40.31, "lon": 36.55},
]

sehir_df = pd.DataFrame(sehirler)
print(f"Toplam sehir sayisi: {len(sehir_df)}")
print(sehir_df)
```

```python
# Her sehir icin hava verisi cek
tum_hava_verisi = []

for _, satir in sehir_df.iterrows():
    print(f"{satir['sehir']} icin veri cekiliyor...")
    
    url = "https://archive-api.open-meteo.com/v1/archive"
    params = {
        "latitude": satir["lat"],
        "longitude": satir["lon"],
        "start_date": "2022-01-01",
        "end_date": "2024-12-31",
        "daily": [
            "temperature_2m_max",
            "temperature_2m_min",
            "precipitation_sum",
            "windspeed_10m_max"
        ],
        "timezone": "Europe/Istanbul"
    }
    
    response = requests.get(url, params=params)
    data = response.json()
    
    df_temp = pd.DataFrame(data["daily"])
    df_temp["sehir"] = satir["sehir"]
    df_temp["bolge"] = satir["bolge"]
    
    tum_hava_verisi.append(df_temp)

hava_df = pd.concat(tum_hava_verisi, ignore_index=True)
hava_df.rename(columns={
    "time": "tarih",
    "temperature_2m_max": "max_sicaklik",
    "temperature_2m_min": "min_sicaklik",
    "precipitation_sum": "yagis",
    "windspeed_10m_max": "ruzgar_hizi"
}, inplace=True)

print(f"\nToplam satir: {len(hava_df)}")
print(f"Sutunlar: {hava_df.columns.tolist()}")
print(hava_df.head())
```

```python
# Veriyi kaydet
hava_df.to_csv("../data/turkiye_hava_verisi.csv", index=False)
sehir_df.to_csv("../data/sehir_listesi.csv", index=False)

print("Kaydedildi!")
print(f"  → data/turkiye_hava_verisi.csv  ({len(hava_df)} satir)")
print(f"  → data/sehir_listesi.csv        ({len(sehir_df)} satir)")
```
